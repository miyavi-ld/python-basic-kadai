```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}さん ({self.age}歳) は大人です。")
        else:
            print(f"{self.name}さん ({self.age}歳) は大人ではありません。")
human1 = Human("太郎", 25)
human2 = Human("花子", 17)
human3 = Human("次郎", 36)
human4 = Human("三郎", 13)

humans_list = [human1, human2, human3, human4]
for person in humans_list:
    person.check_adult()
print("--- 年齢判定終了 ---")
```
