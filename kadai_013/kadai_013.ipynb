```python
def calculate_total_price(price, tax_rate_percent):
    tax_rate = tax_rate_percent / 100
    tax_amount = price * tax_rate
    total_price = price + tax_amount
    return total_price
item_price = 110
tax_rate = 10
result = calculate_total_price(item_price, tax_rate)

print(f"商品金額: {item_price} 円")
print(f"消費税率: {tax_rate} %")
print(f"消費税込みの合計金額: {result} 円")
```
